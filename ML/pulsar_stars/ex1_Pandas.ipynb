```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
#type(scaler)
```

```python
#stars_scaled = scaler.fit_transform(all_stars[['MIP']])
stars_scaled = scaler.fit_transform(all_stars)
stars_scaled
```

```python
list_columns = list(all_stars.columns)
norm_stars = pd.DataFrame(stars_scaled, columns = list_columns)
norm_stars.describe()
```

```python
star = [0.337,0.513,0.425, 0.036,0.928,0.721, 0.714,0.261]
```

```python
evklid_metric_list = []
for i in range(len(norm_stars)):
    s1 = norm_stars.loc[i]
    evklid_metric = 0
    for j in range(len(s1)):
        evklid_metric+=(s1[j]-star[j])**2
    evklid_metric_list.append(evklid_metric**0.5)

```
